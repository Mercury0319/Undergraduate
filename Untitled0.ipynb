```python
# Import necessary libraries
import pandas as pd # For data manipulation and analysis
import matplotlib.pyplot as plt # For plotting graphs
import matplotlib.dates as mdates # For formatting dates on the plot axis
import json # To parse JSON-like strings in the 'visits_by_day' column
import math # For calculating number of plot rows/cols

# --- Configuration ---
# List of the three CSV files to be processed.
# Assumes each file covers one consecutive week.
files_to_process = [
    'sc_pickens_feb2020_merged_1.csv', # Expected: Week Before (Jan 27 - Feb 2)
    'sc_pickens_feb2020_merged_0.csv',   # Expected: Week During (Feb 3 - Feb 9)
    'sc_pickens_feb2020_merged_2.csv'  # Expected: Week After (Feb 10 - Feb 16)
]
# Define the key column names from the SafeGraph data structure
date_col = 'sg_wp__date_range_start' # Column indicating the start date of the 7-day record
visits_col = 'sg_wp__visits_by_day' # Column containing the string list of 7 daily visit counts
location_cols = ['placekey', 'sg_wp__location_name'] # Columns used to identify the Point of Interest (POI)
category_col = 'sg_c__top_category' # Optional column for POI category analysis
N_TOP_CATEGORIES = 5 # Number of top categories to plot individually

# --- Function to Process Each SafeGraph File ---
def process_safegraph_file(filename):
    """
    Loads a SafeGraph pattern file specified by filename.
    Parses the 'sg_wp__visits_by_day' column, assuming it's a list of 7 daily counts
    starting from the date specified in 'sg_wp__date_range_start'.
    Extracts daily visit data for each POI.
    Returns a pandas DataFrame with columns:
        placekey, location_name, date, visits, category (if available).
    Returns None if the file cannot be loaded, essential columns are missing,
    or no valid data can be extracted.

    Args:
        filename (str): The path to the CSV file to process.

    Returns:
        pandas.DataFrame or None: A DataFrame containing daily visit data,
                                  or None if processing fails.
    """
    print(f"\n--- Processing file: {filename} ---")
    try:
        # Load the CSV file into a pandas DataFrame
        df = pd.read_csv(filename)
        print(f"Loaded successfully. Shape: {df.shape}")
    except FileNotFoundError:
        print(f"Error: File not found - {filename}")
        return None
    except Exception as e:
        # Catch any other potential loading errors
        print(f"Error loading {filename}: {e}")
        return None

    # Check if all essential columns exist in the loaded DataFrame
    essential_cols = location_cols + [date_col, visits_col]
    if not all(col in df.columns for col in essential_cols):
        print(f"Error: Essential columns ({essential_cols}) missing in {filename}.")
        return None

    # Select relevant columns for processing, including category if present
    # Create a list of columns to keep
    cols_to_select = essential_cols + ([category_col] if category_col in df.columns else [])
    # Create a new DataFrame with only the selected columns to avoid modifying the original
    df_processed = df[cols_to_select].copy()

    # Initialize list to store structured daily visit records
    daily_visits_data = []
    parse_errors = 0 # Counter for rows with parsing issues
    total_rows = len(df_processed)

    # Iterate through each row (each row represents one POI's data for a week)
    for index, row in df_processed.iterrows():
        try:
            # --- Data Validation and Parsing ---
            # Skip row if the visit data column is null/NaN or not a string
            if pd.isna(row[visits_col]) or not isinstance(row[visits_col], str):
                 parse_errors += 1
                 continue # Move to the next row

            # Parse the string representation of the list (e.g., '[1,2,3,4,5,6,7]') into a Python list
            visits_list = json.loads(row[visits_col])

            # Validate that the parsed data is a list and contains exactly 7 elements (for 7 days)
            if isinstance(visits_list, list) and len(visits_list) == 7:
                # --- Date Calculation ---
                # Convert the start date string to a datetime object
                # Use .normalize() to set the time component to 00:00:00 (ensures consistency)
                start_date = pd.to_datetime(row[date_col]).normalize()
                # Generate a sequence of 7 dates starting from the start_date
                dates = pd.date_range(start=start_date, periods=7)

                # --- Record Creation ---
                # Create a separate record (dictionary) for each day within the 7-day period
                for i, visit_count in enumerate(visits_list):
                    record = {
                        'placekey': row['placekey'],
                        # Use .get() for safer access in case the column is missing somehow
                        'location_name': row.get('sg_wp__location_name', 'N/A'),
                        'date': dates[i], # Assign the specific date for this day's count
                        'visits': int(visit_count) # Ensure visits are stored as integers
                    }
                    # Add category information if the column exists and the value is not null
                    # Handle potential NaN values in category column before assigning
                    cat_value = 'Unknown' # Default category
                    if category_col in row and pd.notna(row[category_col]):
                         cat_value = row[category_col]
                    record['category'] = cat_value

                    # Add the daily record to our list
                    daily_visits_data.append(record)
            else:
                 # Increment error count if data is not a list of 7
                 parse_errors += 1
        except (json.JSONDecodeError, TypeError, ValueError) as e:
             # Handle errors during JSON parsing (malformed string) or int conversion (non-numeric visits)
             print(f"Warning: Error processing row {index}: {e} - Data: {row[visits_col]}")
             parse_errors += 1

    # Report any parsing issues encountered for this file
    if parse_errors > 0:
        print(f"Encountered parsing issues or unexpected format in {parse_errors} out of {total_rows} rows.")

    # Return a DataFrame created from the list of daily records if data was extracted
    if daily_visits_data:
        return pd.DataFrame(daily_visits_data)
    else:
        # Return None if no valid data could be extracted from the file
        print("No valid data extracted from this file.")
        return None

# --- Main Execution Block ---

# Initialize an empty list to hold the DataFrames generated from each file
all_daily_data = []

# Loop through the list of filenames to process
for f in files_to_process:
    # Call the processing function for the current file
    df_daily = process_safegraph_file(f)
    # Only add the resulting DataFrame to the list if processing was successful (returned a DataFrame)
    if df_daily is not None:
        all_daily_data.append(df_daily)

# --- Combine and Analyze Data (only if all files were processed successfully) ---
if len(all_daily_data) == len(files_to_process):
    print("\n--- Combining Data from All Weeks ---")
    # Concatenate the list of individual weekly DataFrames into one large DataFrame
    # ignore_index=True creates a new continuous index for the combined DataFrame
    df_combined_daily = pd.concat(all_daily_data, ignore_index=True)
    print(f"Combined daily DataFrame shape: {df_combined_daily.shape}")

    # --- Aggregate Total Visits Per Day (Overall) ---
    print("\n--- Aggregating Total Daily Visits (Overall - 3 Weeks) ---")
    # Group the combined data by 'date' and calculate the sum of 'visits' for each date
    daily_totals_3wk = df_combined_daily.groupby('date')['visits'].sum().reset_index()
    # Ensure the aggregated data is sorted chronologically by date for plotting
    daily_totals_3wk = daily_totals_3wk.sort_values(by='date')

    print("\nTotal Daily Visits (Aggregated across all POIs, Jan 27 - Feb 16):")
    print(daily_totals_3wk) # Print aggregated totals

    # --- Plot Overall Visits ---
    print("\n--- Generating Overall 3-Week Plot ---")
    fig_overall, ax_overall = plt.subplots(figsize=(15, 7))
    ax_overall.plot(daily_totals_3wk['date'], daily_totals_3wk['visits'], marker='o', linestyle='-')
    flood_date = pd.to_datetime('2020-02-06')
    ax_overall.axvline(flood_date, color='red', linestyle='--', linewidth=2, label='Feb 6 (Flood Event)')
    week_starts = [pd.to_datetime('2020-01-27'), pd.to_datetime('2020-02-03'), pd.to_datetime('2020-02-10')]
    for start_date in week_starts:
        ax_overall.axvline(start_date, color='grey', linestyle=':', linewidth=1, alpha=0.7)
    ax_overall.set_title('Overall Total Daily Visits to POIs (Jan 27 - Feb 16, 2020)')
    ax_overall.set_xlabel('Date')
    ax_overall.set_ylabel('Total Recorded Visits')
    ax_overall.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d (%a)'))
    ax_overall.xaxis.set_major_locator(mdates.DayLocator(interval=1))
    plt.setp(ax_overall.get_xticklabels(), rotation=60, ha='right') # Use plt.setp for clarity
    ax_overall.grid(True, axis='y', linestyle=':', alpha=0.7)
    ax_overall.legend()
    plt.tight_layout()
    plt.show()
    print("Overall 3-Week Plot generated.")


    # --- Category Analysis ---
    print("\n--- Analyzing Visits by Top Category ---")

    # Calculate total visits per category over the 3 weeks
    category_totals = df_combined_daily.groupby('category')['visits'].sum().sort_values(ascending=False)

    # Identify the top N categories (excluding 'Unknown' if desired, but keep for now)
    top_categories = category_totals.head(N_TOP_CATEGORIES).index.tolist()
    print(f"\nTop {N_TOP_CATEGORIES} Categories by Total Visits:")
    print(category_totals.head(N_TOP_CATEGORIES))

    # Filter the combined daily data to include only the top categories
    df_top_categories = df_combined_daily[df_combined_daily['category'].isin(top_categories)].copy()

    # Aggregate daily visits for these top categories
    daily_category_totals = df_top_categories.groupby(['date', 'category'])['visits'].sum().reset_index()
    daily_category_totals = daily_category_totals.sort_values(by=['category', 'date']) # Sort for consistent plotting

    # --- Plot Visits for Top Categories ---
    print(f"\n--- Generating Plots for Top {N_TOP_CATEGORIES} Categories ---")

    # Determine grid size for subplots
    n_cols = 2
    n_rows = math.ceil(N_TOP_CATEGORIES / n_cols)
    fig_cat, axes_cat = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 5), sharex=True) # Share X axis
    axes_cat = axes_cat.flatten() # Flatten the 2D array of axes for easy iteration

    # Define the flood date once
    flood_date = pd.to_datetime('2020-02-06')

    # Plot each top category on a separate subplot
    for i, category_name in enumerate(top_categories):
        ax = axes_cat[i]
        # Filter data for the current category
        category_data = daily_category_totals[daily_category_totals['category'] == category_name]

        # Plot the data
        ax.plot(category_data['date'], category_data['visits'], marker='.', linestyle='-', label=category_name)

        # Add flood line and week separators
        ax.axvline(flood_date, color='red', linestyle='--', linewidth=1.5, label='Feb 6')
        for start_date in week_starts:
            ax.axvline(start_date, color='grey', linestyle=':', linewidth=1, alpha=0.7)

        # Formatting
        ax.set_title(f'Daily Visits: {category_name}')
        ax.set_ylabel('Recorded Visits')
        ax.grid(True, axis='y', linestyle=':', alpha=0.7)
        # ax.legend() # Optional: legend can make plots busy, title is clear

        # Format x-axis for the bottom plots only
        if i >= (n_rows - 1) * n_cols: # Check if it's in the last row
             ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d (%a)'))
             ax.xaxis.set_major_locator(mdates.DayLocator(interval=2)) # Show ticks every 2 days
             plt.setp(ax.get_xticklabels(), rotation=45, ha='right')
        else:
            plt.setp(ax.get_xticklabels(), visible=False) # Hide x-labels for upper plots

    # Hide any unused subplots if N_TOP_CATEGORIES is not a multiple of n_cols
    for j in range(i + 1, len(axes_cat)):
        axes_cat[j].set_visible(False)

    # Add a main title and adjust layout
    fig_cat.suptitle(f'Daily Visits for Top {N_TOP_CATEGORIES} POI Categories (Jan 27 - Feb 16, 2020)', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.97]) # Adjust layout to make room for suptitle
    plt.show()
    print(f"Plots for Top {N_TOP_CATEGORIES} Categories generated.")


else:
    # This message is shown if any of the files failed to load or process
    print("\nCould not process data from all required files. Combined analysis aborted.")

```

```python
# Import necessary libraries
import pandas as pd # For data manipulation and analysis
import matplotlib.pyplot as plt # For plotting graphs
import matplotlib.dates as mdates # For formatting dates on the plot axis
import json # To parse JSON-like strings
import math # For calculating number of plot rows/cols

# --- Configuration ---
# UPDATED List of the three CSV files to be processed based on user input.
# Ensure these exact filenames correspond to the uploaded files.
files_to_process = [
    'sc_pickens_feb2020_merged_1.csv', # Expected: Week Before (Jan 27 - Feb 2)
    'sc_pickens_feb2020_merged_0.csv', # Expected: Week During (Feb 3 - Feb 9) - NOTE: User provided _0, original was (2)
    'sc_pickens_feb2020_merged_2.csv'  # Expected: Week After (Feb 10 - Feb 16) - NOTE: User provided _2, original was _2 (1)
]
# Define the key column names from the SafeGraph data structure
date_col = 'sg_wp__date_range_start' # Column indicating the start date of the 7-day record
hourly_visits_col = 'sg_wp__visits_by_each_hour' # Column with the list of 168 hourly visit counts (7 days * 24 hours)
location_cols = ['placekey', 'sg_wp__location_name'] # Columns used to identify the Point of Interest (POI)
category_col = 'sg_c__top_category' # Optional column for POI category analysis

# Define the specific dates (Thursdays) to compare
thursday_before = pd.to_datetime('2020-01-30').normalize()
thursday_during = pd.to_datetime('2020-02-06').normalize() # Flood event day
thursday_after = pd.to_datetime('2020-02-13').normalize()
dates_to_compare = [thursday_before, thursday_during, thursday_after]

# --- Function to Process Each SafeGraph File for HOURLY Data ---
def process_safegraph_file_hourly(filename):
    """
    Loads a SafeGraph pattern file specified by filename.
    Parses the 'sg_wp__visits_by_each_hour' column, assuming it's a list of 168 hourly counts
    starting from the date specified in 'sg_wp__date_range_start'.
    Extracts hourly visit data for each POI.
    Returns a pandas DataFrame with columns:
        placekey, location_name, datetime, hour, date, visits, category (if available).
    Returns None if the file cannot be loaded, essential columns are missing,
    or no valid data can be extracted.

    Args:
        filename (str): The path to the CSV file to process.

    Returns:
        pandas.DataFrame or None: A DataFrame containing hourly visit data,
                                  or None if processing fails.
    """
    print(f"\n--- Processing file for HOURLY data: {filename} ---")
    try:
        # NOTE: Assuming the filenames provided by the user directly map to the uploaded files.
        # If they are different, this pd.read_csv step will fail.
        df = pd.read_csv(filename)
        print(f"Loaded successfully. Shape: {df.shape}")
    except FileNotFoundError:
         print(f"Error: File not found - {filename}. Please ensure this filename matches an uploaded file.")
         return None
    except Exception as e:
        print(f"Error loading {filename}: {e}")
        return None

    # Check if essential columns exist (including the hourly visits column)
    essential_cols = location_cols + [date_col, hourly_visits_col]
    if not all(col in df.columns for col in essential_cols):
        print(f"Error: Essential columns for hourly analysis missing in {filename}.")
        return None

    # Select relevant columns
    cols_to_select = essential_cols + ([category_col] if category_col in df.columns else [])
    df_processed = df[cols_to_select].copy()

    hourly_visits_data = []
    parse_errors = 0
    total_rows = len(df_processed)

    for index, row in df_processed.iterrows():
        try:
            if pd.isna(row[hourly_visits_col]) or not isinstance(row[hourly_visits_col], str):
                 parse_errors += 1
                 continue
            # Parse the string list of 168 hourly visits
            visits_list_hourly = json.loads(row[hourly_visits_col])

            # Validate that it's a list of exactly 168 hours
            if isinstance(visits_list_hourly, list) and len(visits_list_hourly) == 168:
                # Determine the start datetime (keep time info initially for freq='H')
                start_datetime = pd.to_datetime(row[date_col])
                # Generate timestamps for each hour over the 7 days (168 periods)
                timestamps = pd.date_range(start=start_datetime, periods=168, freq='H')

                # Create a record for each hour
                for i, visit_count in enumerate(visits_list_hourly):
                    ts = timestamps[i]
                    record = {
                        'placekey': row['placekey'],
                        'location_name': row.get('sg_wp__location_name', 'N/A'),
                        'datetime': ts, # Full timestamp
                        'date': ts.normalize(), # Date part only
                        'hour': ts.hour, # Hour of the day (0-23)
                        'visits': int(visit_count)
                    }
                    # Add category information if available
                    cat_value = 'Unknown'
                    if category_col in row and pd.notna(row[category_col]):
                         cat_value = row[category_col]
                    record['category'] = cat_value
                    hourly_visits_data.append(record)
            else:
                 parse_errors += 1
        except (json.JSONDecodeError, TypeError, ValueError) as e:
             print(f"Warning: Error processing row {index} for hourly data: {e}")
             parse_errors += 1

    if parse_errors > 0:
        print(f"Encountered parsing issues in {parse_errors} out of {total_rows} rows for hourly data.")

    if hourly_visits_data:
        return pd.DataFrame(hourly_visits_data)
    else:
        print("No valid hourly data extracted from this file.")
        return None

# --- Main Execution Block for Hourly Analysis ---

# Process All Files for Hourly Data
all_hourly_data = []
for f in files_to_process:
    df_hourly = process_safegraph_file_hourly(f)
    if df_hourly is not None:
        all_hourly_data.append(df_hourly)

# --- Combine and Analyze Hourly Data (only if all files were processed) ---
if len(all_hourly_data) == len(files_to_process):
    print("\n--- Combining HOURLY Data from All Weeks ---")
    df_combined_hourly = pd.concat(all_hourly_data, ignore_index=True)
    print(f"Combined HOURLY DataFrame shape: {df_combined_hourly.shape}")

    # --- Filter for Specific Dates ---
    print(f"\n--- Filtering for dates: {[d.strftime('%Y-%m-%d') for d in dates_to_compare]} ---")
    # Ensure the 'date' column is timezone-naive for comparison if dates_to_compare are naive
    if df_combined_hourly['date'].dt.tz is not None:
        print("Converting 'date' column to timezone-naive for filtering.")
        df_combined_hourly['date'] = df_combined_hourly['date'].dt.tz_localize(None)

    df_filtered_hourly = df_combined_hourly[df_combined_hourly['date'].isin(dates_to_compare)].copy()
    print(f"Filtered HOURLY DataFrame shape: {df_filtered_hourly.shape}")

    if df_filtered_hourly.empty:
        print("No data found for the specified comparison dates. Cannot generate plot.")
    else:
        # --- Aggregate Visits per Hour for Filtered Dates ---
        print("\n--- Aggregating Visits per Hour for Comparison Dates ---")
        hourly_totals_compare = df_filtered_hourly.groupby(['date', 'hour'])['visits'].sum().reset_index()
        print(hourly_totals_compare.head()) # Show sample of aggregated data

        # --- Plot Hourly Comparison ---
        print("\n--- Generating Hourly Comparison Plot ---")
        fig_hourly, ax_hourly = plt.subplots(figsize=(15, 7))

        # Plot each comparison date separately
        for dt in dates_to_compare:
            data_to_plot = hourly_totals_compare[hourly_totals_compare['date'] == dt]
            label_str = dt.strftime('%Y-%m-%d (%a)')
            if dt == thursday_during:
                label_str += " (Flood Event)"
                # Emphasize the flood event day line
                ax_hourly.plot(data_to_plot['hour'], data_to_plot['visits'], marker='o', linestyle='-', linewidth=2.5, zorder=5, label=label_str)
            else:
                # Plot baseline days with lighter style
                ax_hourly.plot(data_to_plot['hour'], data_to_plot['visits'], marker='.', linestyle='--', alpha=0.8, label=label_str)

        # Formatting the plot
        ax_hourly.set_title('Comparison of Hourly Visits on Specific Thursdays')
        ax_hourly.set_xlabel('Hour of Day (0-23)')
        ax_hourly.set_ylabel('Total Recorded Visits')
        ax_hourly.set_xticks(range(0, 24)) # Set ticks for every hour
        ax_hourly.grid(True, axis='both', linestyle=':', alpha=0.6) # Grid on both axes
        ax_hourly.legend(title="Date")
        plt.tight_layout()
        plt.show()
        print("Hourly Comparison Plot generated.")

else:
    print("\nCould not process HOURLY data from all required files. Hourly analysis aborted.")
    print("Please ensure the filenames in 'files_to_process' exactly match the uploaded files.")

```

```python
# Import necessary libraries
import pandas as pd # For data manipulation and analysis
import json # To parse JSON-like strings
from collections import defaultdict # For easier aggregation

# --- Configuration ---
# UPDATED List of the three CSV files based on user input.
# !!! IMPORTANT: Ensure these filenames exactly match the uploaded files !!!
files_to_process = [
    'sc_pickens_feb2020_merged_1.csv', # Expected: Week Before (Jan 27 - Feb 2)
    'sc_pickens_feb2020_merged_0.csv', # Expected: Week During (Feb 3 - Feb 9)
    'sc_pickens_feb2020_merged_2.csv'  # Expected: Week After (Feb 10 - Feb 16)
]
# Define the key column names from the SafeGraph data structure
date_col = 'sg_wp__date_range_start' # Column indicating the start date of the 7-day record
cbg_col = 'sg_wp__visitor_home_cbgs' # Column with the visitor home CBG JSON
location_cols = ['placekey'] # Need placekey to link, location name is optional here

# --- Function to Process Each SafeGraph File for SPATIAL (CBG) Data ---
def process_safegraph_file_spatial(filename):
    """
    Loads a SafeGraph pattern file specified by filename.
    Parses the 'sg_wp__visitor_home_cbgs' column (JSON string mapping CBG to count).
    Extracts weekly visitor counts originating from each CBG for each POI.
    Returns a pandas DataFrame with columns:
        placekey, week_start_date, home_cbg, weekly_visitors_from_cbg.
    Returns None if the file cannot be loaded, essential columns are missing,
    or no valid data can be extracted.

    Args:
        filename (str): The path to the CSV file to process.

    Returns:
        pandas.DataFrame or None: A DataFrame containing weekly visitor counts by home CBG,
                                  or None if processing fails.
    """
    print(f"\n--- Processing file for SPATIAL (CBG) data: {filename} ---")
    try:
        # Load the CSV file
        df = pd.read_csv(filename)
        print(f"Loaded successfully. Shape: {df.shape}")
    except FileNotFoundError:
         print(f"Error: File not found - {filename}. Please ensure this filename matches an uploaded file.")
         return None
    except Exception as e:
        print(f"Error loading {filename}: {e}")
        return None

    # Check if essential columns exist
    essential_cols = location_cols + [date_col, cbg_col]
    if not all(col in df.columns for col in essential_cols):
        print(f"Error: Essential columns for spatial analysis missing in {filename}.")
        return None

    # Select relevant columns
    cols_to_select = essential_cols
    df_processed = df[cols_to_select].copy()

    # List to store extracted records
    cbg_visits_data = []
    parse_errors = 0
    total_rows = len(df_processed)

    # Iterate through each row (POI's weekly data)
    for index, row in df_processed.iterrows():
        try:
            # Get the week start date (normalize to remove time)
            week_start_date = pd.to_datetime(row[date_col]).normalize()
            placekey = row['placekey']

            # Check if CBG data is present and is a string
            if pd.isna(row[cbg_col]) or not isinstance(row[cbg_col], str):
                 parse_errors += 1
                 continue # Skip if CBG data is missing or not a string

            # Parse the JSON string (e.g., '{"cbg1": count1, "cbg2": count2}')
            cbg_dict = json.loads(row[cbg_col])

            # Check if parsing resulted in a dictionary
            if isinstance(cbg_dict, dict):
                # Iterate through each CBG and its visitor count in the dictionary
                for cbg_fips, visitor_count in cbg_dict.items():
                    # Basic validation: Check if CBG looks like a FIPS code (string of digits)
                    # and count is a non-negative integer
                    if isinstance(cbg_fips, str) and cbg_fips.isdigit() and isinstance(visitor_count, int) and visitor_count >= 0:
                        record = {
                            'placekey': placekey,
                            'week_start_date': week_start_date,
                            'home_cbg': cbg_fips, # Store the CBG FIPS code
                            'weekly_visitors_from_cbg': visitor_count # Store the count for this week
                        }
                        cbg_visits_data.append(record)
                    else:
                        # Increment error count if CBG or count format is unexpected
                        # print(f"Warning: Unexpected format in CBG dict item: CBG={cbg_fips}, Count={visitor_count} in row {index}")
                        parse_errors += 1 # Count this as a partial error within the row
            else:
                # Increment error count if the parsed JSON is not a dictionary
                parse_errors += 1

        except (json.JSONDecodeError, TypeError, ValueError) as e:
             # Handle errors during JSON parsing itself
             print(f"Warning: Error parsing CBG JSON in row {index}: {e} - Data: {row[cbg_col]}")
             parse_errors += 1

    if parse_errors > 0:
        print(f"Encountered parsing issues or unexpected format in {parse_errors} entries/rows for CBG data.")

    if cbg_visits_data:
        return pd.DataFrame(cbg_visits_data)
    else:
        print("No valid CBG visitor data extracted from this file.")
        return None

# --- Main Execution Block for Spatial (CBG) Analysis ---

# Process All Files for CBG Data
all_cbg_data = []
for f in files_to_process:
    df_cbg = process_safegraph_file_spatial(f)
    if df_cbg is not None:
        all_cbg_data.append(df_cbg)

# --- Combine and Analyze CBG Data (only if all files were processed) ---
if len(all_cbg_data) == len(files_to_process):
    print("\n--- Combining Weekly CBG Visitor Data from All Weeks ---")
    df_combined_cbg = pd.concat(all_cbg_data, ignore_index=True)
    print(f"Combined CBG DataFrame shape: {df_combined_cbg.shape}")

    # --- Aggregate Total Visitors from each CBG per Week ---
    # Sum visitors across all POIs for each home_cbg and week_start_date combination
    print("\n--- Aggregating Total Weekly Visitors per Home CBG ---")
    weekly_cbg_totals = df_combined_cbg.groupby(['week_start_date', 'home_cbg'])['weekly_visitors_from_cbg'].sum().reset_index()
    print(f"Aggregated weekly CBG totals shape: {weekly_cbg_totals.shape}")
    print(weekly_cbg_totals.head())

    # --- Pivot Data for Comparison ---
    print("\n--- Pivoting Data for Week-over-Week Comparison ---")
    # Pivot the table to have weeks as columns for easy comparison
    df_pivot = weekly_cbg_totals.pivot(index='home_cbg', columns='week_start_date', values='weekly_visitors_from_cbg').reset_index()

    # Rename columns for clarity - ASSUMES the files correspond to these exact start dates
    # Get the actual date columns from the pivot table
    date_columns = df_pivot.columns[1:] # Exclude 'home_cbg'
    # Sort date columns to ensure correct order
    date_columns_sorted = sorted(date_columns)

    if len(date_columns_sorted) == 3:
        week_before_date = date_columns_sorted[0]
        week_during_date = date_columns_sorted[1]
        week_after_date = date_columns_sorted[2]

        df_pivot = df_pivot.rename(columns={
            week_before_date: 'visits_week_before',
            week_during_date: 'visits_week_during',
            week_after_date: 'visits_week_after'
        })

        # Fill NaN values with 0 (if a CBG had no visitors in a given week)
        df_pivot = df_pivot.fillna(0)

        # --- Calculate Change During Flood Week ---
        print("\n--- Calculating Change in Visitors During Flood Week ---")
        # Calculate absolute change
        df_pivot['change_abs_flood_week'] = df_pivot['visits_week_during'] - df_pivot['visits_week_before']
        # Calculate percentage change (handle division by zero)
        df_pivot['change_pct_flood_week'] = (df_pivot['change_abs_flood_week'] / df_pivot['visits_week_before'].replace(0, pd.NA)) * 100
        # Fill NaN percentages (from division by zero) with 0 or another indicator if preferred
        df_pivot['change_pct_flood_week'] = df_pivot['change_pct_flood_week'].fillna(0)


        # --- Display Results ---
        print("\n--- Comparison of Weekly Visitors by Home CBG ---")
        # Sort by the largest decrease during the flood week
        df_comparison = df_pivot.sort_values(by='change_abs_flood_week', ascending=True)
        print(df_comparison.head(15)) # Show top 15 CBGs with largest decrease

        print("\n--- CBGs with Largest Percentage Decrease ---")
        # Filter out CBGs with low baseline visits before sorting by percentage
        df_comparison_pct = df_pivot[df_pivot['visits_week_before'] > 5].sort_values(by='change_pct_flood_week', ascending=True)
        print(df_comparison_pct.head(15))


        print("\n--- Next Steps for Spatial Visualization ---")
        print("1. Obtain Census Block Group (CBG) geographic boundary data (e.g., Shapefile).")
        print("2. Use libraries like GeoPandas to join this table (df_comparison) with the CBG shapes using 'home_cbg' (FIPS code).")
        print("3. Create choropleth maps visualizing 'visits_week_during' or 'change_abs_flood_week'.")

    else:
        print("Error: Pivot table does not have exactly 3 date columns as expected.")
        print("Pivot table columns:", df_pivot.columns)


else:
    print("\nCould not process CBG data from all required files. Spatial analysis aborted.")
    print("Please ensure the filenames in 'files_to_process' exactly match the uploaded files.")

```
